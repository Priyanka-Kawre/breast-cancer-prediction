```python
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
```

```python
# Reload your dataset
df = pd.read_csv("Breast Cancer METABRIC.csv")

# Standardize text in the 'Overall Survival Status' column
df['Overall Survival Status'] = df['Overall Survival Status'].str.strip().str.upper()
```

```python
# Keep only relevant columns
df_survival = df[['Overall Survival (Months)', 'Overall Survival Status']].copy()

# Drop missing values
df_survival.dropna(inplace=True)

# Remove rows with invalid survival status
df_survival = df_survival[df_survival['Overall Survival Status'].isin(['DECEASED', 'LIVING'])]

# Convert 'Overall Survival (Months)' to numeric
df_survival['Overall Survival (Months)'] = pd.to_numeric(df_survival['Overall Survival (Months)'], errors='coerce')

# Drop again just in case
df_survival.dropna(inplace=True)

# Define duration (T) and event (E)
T = df_survival['Overall Survival (Months)']
E = df_survival['Overall Survival Status'].apply(lambda x: 1 if x == 'DECEASED' else 0)
```

```python
kmf = KaplanMeierFitter()

kmf.fit(durations=T, event_observed=E)
kmf.plot_survival_function()

plt.title("Kaplan-Meier Survival Curve")
plt.xlabel("Time (Months)")
plt.ylabel("Survival Probability")
plt.grid(True)
plt.show()
```

```python
# Filter out missing ER values
df_er = df[['ER Status', 'Overall Survival (Months)', 'Overall Survival Status']].dropna()
df_er['Overall Survival Status'] = df_er['Overall Survival Status'].str.strip().str.upper()

# Convert time and event
T = df_er['Overall Survival (Months)']
E = df_er['Overall Survival Status'].apply(lambda x: 1 if x == 'DECEASED' else 0)

# Plot survival for each ER status
plt.figure(figsize=(8,6))
kmf = KaplanMeierFitter()

for status in df_er['ER Status'].unique():
    ix = df_er['ER Status'] == status
    kmf.fit(T[ix], E[ix], label=str(status))
    kmf.plot_survival_function()

plt.title("Kaplan-Meier Survival by ER Status")
plt.xlabel("Time (Months)")
plt.ylabel("Survival Probability")
plt.legend(title="ER Status")
plt.grid(True)
plt.show()
```

```python
# Clean and filter
df_ht = df[['Hormone Therapy', 'Overall Survival (Months)', 'Overall Survival Status']].dropna()
df_ht['Overall Survival Status'] = df_ht['Overall Survival Status'].str.strip().str.upper()

# Time and event
T = df_ht['Overall Survival (Months)']
E = df_ht['Overall Survival Status'].apply(lambda x: 1 if x == 'DECEASED' else 0)

# Plot
plt.figure(figsize=(8,6))
kmf = KaplanMeierFitter()

for group in df_ht['Hormone Therapy'].unique():
    ix = df_ht['Hormone Therapy'] == group
    kmf.fit(T[ix], E[ix], label=str(group))
    kmf.plot_survival_function()

plt.title("Survival by Hormone Therapy")
plt.xlabel("Time (Months)")
plt.ylabel("Survival Probability")
plt.legend(title="Hormone Therapy")
plt.grid(True)
plt.show()
```

```python
# Import libraries
import pandas as pd
from lifelines import CoxPHFitter

# Load data
df = pd.read_csv("Breast Cancer METABRIC.csv")
df['Overall Survival Status'] = df['Overall Survival Status'].str.strip().str.upper()

# Prepare data
cox_df = df[['Overall Survival (Months)', 'Overall Survival Status',
             'Age at Diagnosis', 'Tumor Size', 'Neoplasm Histologic Grade',
             'ER Status', 'PR Status', 'HER2 Status', 'Hormone Therapy',
             'Radio Therapy', 'Tumor Stage']].copy()

cox_df.dropna(inplace=True)
cox_df = cox_df[cox_df['Overall Survival Status'].isin(['DECEASED', 'LIVING'])]
cox_df['Event'] = cox_df['Overall Survival Status'].apply(lambda x: 1 if x == 'DECEASED' else 0)

# Encode categorical variables
cox_df = pd.get_dummies(cox_df, drop_first=True)

# Drop problematic column if it appears
cox_df.drop(columns=['Overall Survival Status_LIVING'], errors='ignore', inplace=True)

# Fit the model
cph = CoxPHFitter()
cph.fit(cox_df, duration_col='Overall Survival (Months)', event_col='Event')

# Show summary
cph.print_summary()

```

```python
# Standardize survival status column
df['Overall Survival Status'] = df['Overall Survival Status'].str.strip().str.upper()

# Create the target column: 1 = survived ≥10 years, 0 = died <10 years
df['Ten_Year_Mortality'] = df.apply(
    lambda x: 0 if x['Overall Survival Status'] == 'DECEASED' and x['Overall Survival (Months)'] < 120 else 1,
    axis=1
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import seaborn as sns
import matplotlib.pyplot as plt

# Feature selection (update as needed)
features = ['Age at Diagnosis', 'Tumor Size', 'Neoplasm Histologic Grade',
            'ER Status', 'PR Status', 'HER2 Status', 'Hormone Therapy',
            'Radio Therapy', 'Tumor Stage']

df_clf = df[features + ['Ten_Year_Mortality']].dropna()

# Encode categorical features
df_clf = pd.get_dummies(df_clf, drop_first=True)

# Features and target
X = df_clf.drop('Ten_Year_Mortality', axis=1)
y = df_clf['Ten_Year_Mortality']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_scaled, y_train)
y_pred_log = logreg.predict(X_test_scaled)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print("🔹 Logistic Regression:\n", classification_report(y_test, y_pred_log))
print("🔹 Random Forest:\n", classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check class distribution
print("After SMOTE:", y_train_smote.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_smote, y_train_smote)

y_pred_log = logreg.predict(X_test_scaled)

from sklearn.metrics import classification_report, roc_auc_score

print("🔹 Logistic Regression Report:\n", classification_report(y_test, y_pred_log))
print("🔹 ROC-AUC (Logistic Regression):", roc_auc_score(y_test, y_pred_log))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_smote, y_train_smote)

y_pred_rf = rf.predict(X_test)

print("🔹 Random Forest Report:\n", classification_report(y_test, y_pred_rf))
print("🔹 ROC-AUC (Random Forest):", roc_auc_score(y_test, y_pred_rf))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Initialize and train
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train_smote, y_train_smote)

# Predict
y_pred_xgb = xgb.predict(X_test)

# Evaluate
print("🔹 XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
print("🔹 ROC-AUC (XGBoost):", roc_auc_score(y_test, y_pred_xgb))
```
